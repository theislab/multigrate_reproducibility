```python
import pandas as pd
import matplotlib.pyplot as plt
```

## no unimodal RNA in the train data

```python
df = pd.read_csv('cell-systems-no-rna-in-train-wandb_export_2026-01-22T10_58_25.019+01_00.csv', sep=';')
df['Setting'] = df['Name'].str.split('_').str[0] + '_' + df['Name'].str.split('_').str[1]
df = df[df['State'] == 'finished']
columns_to_keep = [
    'test_pearson', 'test_spearman', 'Overall Score', 'Setting',
    'params.alignment_type', 'params.batch_size', 'params.cond_dim', 'params.integ',
    'params.kl', 'params.lr', 'params.mix', 'params.mod_0', 'params.mod_1', 'params.modality_alignment', 'params.query_max_epochs',
]
df = df[columns_to_keep]
df
```

```python
# remove test runs
df = df[df['params.query_max_epochs'] != 2]
df
```

```python
for modality in ['cite', 'multiome']:
    for setting in ['_25', '_50', '_75']:
        df_ = df[df['Setting'].str.contains(modality)]
        df_ = df_[df_['Setting'].str.contains(setting)]
        df_mix = df_[df_['params.mix'] == 'mixture']
        df_prod = df_[df_['params.mix'] == 'product']
        
        print(modality)
        print(setting)
        print('mixture:', df_mix.shape)
        print(df_mix.sort_values('test_pearson', ascending=False).head(1)[['test_pearson', 'test_spearman', 'Overall Score']])
        print(df_mix.sort_values('test_pearson', ascending=False).head(5)[['test_pearson', 'test_spearman', 'Overall Score']].agg(['mean', 'std']))
        print('product:', df_prod.shape)
        print(df_prod.sort_values('test_pearson', ascending=False).head(1)[['test_pearson', 'test_spearman', 'Overall Score']])
        print(df_prod.sort_values('test_pearson', ascending=False).head(5)[['test_pearson', 'test_spearman', 'Overall Score']].agg(['mean', 'std']))
        print('---')
```

```python
for modality in ['cite', 'multiome']:
    for metric in ['test_pearson', 'test_spearman', 'Overall Score']:
        best_mix_values = {}
        best5_mix_values = {}
        best_prod_values = {}
        best5_prod_values = {}
        for setting in ['_25', '_50', '_75']:
            df_ = df[df['Setting'].str.contains(modality)]
            df_ = df_[df_['Setting'].str.contains(setting)]
            df_mix = df_[df_['params.mix'] == 'mixture']
            df_prod = df_[df_['params.mix'] == 'product']

            best_mix_values[setting] = df_mix.sort_values('test_pearson', ascending=False).head(1)
            best5_mix_values[setting] = df_mix.sort_values('test_pearson', ascending=False).head(5)[[metric]].agg(['mean'])

            best_prod_values[setting] = df_prod.sort_values('test_pearson', ascending=False).head(1)
            best5_prod_values[setting] = df_prod.sort_values('test_pearson', ascending=False).head(5)[[metric]].agg(['mean'])

        # line plot per modality, metric
        # settings on x axis
        settings = ['25%', '50%', '75%']
        best_mix_values = [best_mix_values[setting][metric].values[0] for setting in ['_25', '_50', '_75']]
        best5_mix_values = [best5_mix_values[setting].loc['mean', metric] for setting in ['_25', '_50', '_75']]
        best_prod_values = [best_prod_values[setting][metric].values[0] for setting in ['_25', '_50', '_75']]
        best5_prod_values = [best5_prod_values[setting].loc['mean', metric] for setting in ['_25', '_50', '_75']]

        plt.figure(figsize=(6, 4))
        colors = plt.cm.Paired([1, 0]) # colors = plt.cm.Paired([5, 4]) for rna in train
        plt.plot(settings, best_mix_values, marker='o', color=colors[1], label='Best mixture')
        plt.plot(settings, best5_mix_values, marker='o', color=colors[1], linestyle='dashed', label='Top 5 mean mixture')
        plt.plot(settings, best_prod_values, marker='o', color=colors[0], label='Best product')
        plt.plot(settings, best5_prod_values, marker='o', color=colors[0], linestyle='dashed', label='Top 5 mean product')
        
        plt.xlabel('Setting')
        if metric in ['test_pearson', 'test_spearman']:
            metric_name = metric.split('_')[1].capitalize()  + ' correlation'
            plt.ylabel('Correlation')
        else:
            metric_name = 'Overall scIB score'
            plt.ylabel('Score')
        
        if modality == 'cite':
            modality_title = 'CITE-seq'
        else:
            modality_title = 'multiome'

        plt.title(f'{metric_name} by method for the {modality_title} dataset')
        # change x ticks to 25%, 50%, 75%
        plt.xticks(ticks=settings, labels=['25%', '50%', '75%'])
        plt.legend()
        plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5, color='gray', zorder=0)
        plt.savefig(f'./figures/no_rna_in_train_{modality}_{metric}_performance.svg', bbox_inches='tight')
```

```python
df.to_csv('cell-systems-no-rna-in-train_processed.csv', index=False)
```
