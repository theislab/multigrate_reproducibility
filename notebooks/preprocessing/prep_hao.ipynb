```python
cite = sc.read('../data/hao-2020/protein.h5ad')
cite
```

```python
scrna.obs['cell_type'] = scrna.obs['celltype.l2'].astype('category')
cite.obs['cell_type'] = cite.obs['celltype.l2'].astype('category')
```

```python
cite.layers['count']
```

```python
protein_df = pd.DataFrame(index=cite.obs_names)
ref_proteins = cite.var_names
for p in ref_proteins:
    if p.split("-")[-1] == "1" or p.split("-")[-1] == "2":
        root = p.split("-")[0]
        if root not in ["Notch", "TCR"]:
            try:
                protein_df[root] = (
                    cite[:, root + "-1"].layers['count']
                    + cite[:, root + "-2"].layers['count']
                )
                print(root)
            except:
                protein_df[p] = cite[:, p].layers['count']
        else:
            protein_df[p] = cite[:, p].layers['count']
    else:
        protein_df[p] = cite[:, p].layers['count']
protein_df
```

```python
cite_new = sc.AnnData(protein_df.to_numpy())
cite_new.obs = cite.obs.copy()
cite_new
```

```python
cite_new.layers['count'] = cite_new.X.copy()
```

```python
def clr_normalize_each_cell(adata, inplace=True):
    """Normalize count vector for each cell, i.e. for each row of .X"""

    import numpy as np
    import scipy

    def seurat_clr(x):
        # TODO: support sparseness
        s = np.sum(np.log1p(x[x > 0]))
        exp = np.exp(s / len(x))
        return np.log1p(x / exp)

    if not inplace:
        adata = adata.copy()

    # apply to dense or sparse matrix, along axis. returns dense matrix
    adata.X = np.apply_along_axis(
        seurat_clr, 1, (adata.X.A if scipy.sparse.issparse(adata.X) else adata.X)
    )
    return adata
```

```python
cite_new = clr_normalize_each_cell(cite_new)
```
